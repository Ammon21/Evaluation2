```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Optional: load your data here
# df = pd.read_csv("your_file.csv")

# Define the correct custom order
custom_order = ['Unsatisfactory', 'Fair', 'Good', 'Very Good', 'Excellent']
df = df.drop(df.columns[[-1, -2]], axis=1)
# Clean values: strip en-dash, spaces, and standardize
def clean_text(val):
    if pd.isnull(val):
        return val
    return str(val).replace('–', '').strip()

# Clean and encode columns 1 to 19
le = LabelEncoder()
le.classes_ = np.array(custom_order)

for col in df.columns[1:]:
    df[col] = df[col].apply(clean_text)  # Remove en-dash and spaces
    df[col] = df[col].map(lambda x: le.transform([x])[0] + 1 if x in custom_order else np.nan)

# Drop the last two columns by index


# Show the cleaned and encoded DataFrame
df.head()
```
